```python
import requests
import csv
import time

BASE_URL = "https://lunthaelections.com/api/constituency/{}?type=mp&year=2025"
OUTPUT_FILE = "luntha_2025_mp_results.csv"

# CSV columns
columns = [
    "constituency_id", "constituency_name", "district_id",
    "candidate_name", "votes", "percentage", "gender",
    "total_registered", "banned_votes", "votes_cast"
]

all_rows = []

for cid in range(1, 230):   # Constituency IDs 1–229
    url = BASE_URL.format(cid)

    try:
        res = requests.get(url, timeout=10)

        if res.status_code != 200:
            print(f"ID {cid}: No data (status {res.status_code})")
            continue

        data = res.json()

        # Constituency metadata
        cinfo = data.get("constituency", {})
        cname = cinfo.get("name")
        district_id = cinfo.get("district_id")

        # Results for 2025
        results = data.get("resultsByYear", {}).get("2025", [])

        # Stats for 2025
        stats = data.get("constStatsByYear", {}).get("2025", {})
        total_registered = stats.get("total")
        banned = stats.get("banned")
        votes_cast = stats.get("votes")

        if not results:
            print(f"ID {cid}: No candidate data")
            continue

        # Extract each candidate's data
        for r in results:
            row = [
                cid,
                cname,
                district_id,
                r.get("name"),
                r.get("votes"),
                r.get("percentage"),
                r.get("gender"),
                total_registered,
                banned,
                votes_cast
            ]
            all_rows.append(row)

        print(f"✓ Scraped constituency {cid}: {cname}")

        time.sleep(0.3)  # To avoid server overload

    except Exception as e:
        print(f"Error on ID {cid}: {e}")
        continue

# Save to CSV
with open(OUTPUT_FILE, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(columns)
    writer.writerows(all_rows)

print(f"\nDone! Saved {len(all_rows)} rows to {OUTPUT_FILE}")
```

```python
import pandas as pd

df = pd.read_csv("luntha_2025_mp_results.csv")

mask = ["constituency_name", "candidate_name", "votes", "gender"]

df = df[mask]

df['gender'] = df['gender'].replace({0: 'Female', 1: 'Male'})

df = df.sort_values('votes', ascending=False).drop_duplicates(
    subset=['constituency_name'], keep='first'
)

df = df.sort_values('votes', ascending=True)

df.to_excel('scrapped/luntha_cleaned.xlsx')

print(df)
```
